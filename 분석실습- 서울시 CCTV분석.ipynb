## (3) Visualization

### 1) 구별 cctv 수 시각화

```python
data_result["소계"].sort_values().plot(kind="barh", figsize=(7,5), grid = True)
```

### 2) 인구 대비 CCTV 개수를 시각화

#### 인구 수 대비 CCTV 수를 나타내는 "CCTV 비율" 파생변수 추가 : (소계 / 인구수)  *  100

```python
data_result["CCTV비율"] = data_result["소계"]/data_result["인구수"] * 100
data_result.head()
```

```python
data_result["CCTV비율"].sort_values().plot(kind="barh", figsize=(7,5), grid = True)
```

### 3) 인구 수에 따른 CCTV 수 회귀분석 

#### 인구 수에 따른 CCTV 수의 회귀선 구하기

```python
pf = np.polyfit(data_result["인구수"], data_result["소계"], 1)
p1 = np.poly1d(pf)
print(p1)
print("y =", pf[0],"*x +", pf[1])
```

#### 회귀선 시각화

```python
fx = np.linspace(100000, 700000, 100)

plt.figure(figsize = (6, 6))
plt.scatter(data_result["인구수"], data_result["소계"])
plt.plot(fx, pf[0] * fx + pf[1], color = "green")            
plt.xlabel("인구 수")
plt.ylabel("소계")
plt.grid()
```

#### "오차" 파생변수 추가

```python
data_result["오차"] = np.abs(data_result["소계"] - p1(data_result["인구수"]))
data_result.head()
```

```python
result_sort = data_result.sort_values(by="오차", ascending=False)
result_sort
```

#### 오차가 가장 큰 5개 구 표시하여 시각화

```python
fx = np.linspace(100000, 700000, 100)

plt.figure(figsize = (9, 9))
plt.scatter(data_result["인구수"], data_result["소계"], s = 60,    
           c= data_result["오차"])  
plt.plot(fx, pf[0] * fx + pf[1], color = "green") 
plt.xlabel("인구 수")
plt.ylabel("소계")
plt.grid()
plt.colorbar()


##  오차 가장 큰 5개 구 표기

for i in range(5):
    plt.text(result_sort["인구수"][i], result_sort["소계"][i], result_sort.index[i])
```

# 2. 데이터 전처리

- Date 변수 통해서 요일을 뽑아 파생변수(weekdays) 만들기
    - 2016년 1월 1일은 금요일이다.

```python
bike["weekdays"] = ((pd.to_datetime(bike["Date"], format="%d/%m/%Y") - 
     pd.to_datetime("2016-01-01")).dt.days) % 7

bike.head()

bike["weekdays"] = bike["weekdays"].map({0:"금요일", 1:"토요일", 2:"일요일", 
                                         3:"월요일", 4:"화요일", 5:"수요일", 
                                         6:"목요일"})
bike.head()
```

# 3. 시각화

## (1) 시간대 별 평균 대여 수

```python
i = bike.groupby(by=['Hour'], as_index=False).mean()['Rented Bike Count']

type(i)

i.plot.bar()
```

```python
i.plot()

plt.axvline(x=8, color="r")
plt.axvline(x=18, color="r")
plt.text(6, 150, "출근")
plt.text(16, 150, "퇴근")
```

## (2) 요일 별 평균 대여 수

```python
bike.groupby("weekdays").mean()[['Rented Bike Count']]
```

```python
sns.barplot(x="weekdays", y="Rented Bike Count", data=bike, 
           order = ["월요일", "화요일","수요일","목요일","금요일","토요일","일요일"])
```

## (3) 각 요일의 시간대 별 대여 수

```python
_, ax = plt.subplots(2, 4, figsize=(20, 10))

weekdays_list = ["월요일", "화요일", "수요일", "목요일", "금요일", "토요일", "일요일"]

for i in range(7):
    sns.lineplot("Hour", "Rented Bike Count", 
                 data=bike.loc[bike["weekdays"] == weekdays_list[i]],
                ax = ax[i//4, i%4])
    
    ax[i//4, i%4].set_title(weekdays_list[i])
```

## (4) 온도에 따른 대여횟수 

```python
sns.scatterplot(x="Temperature()", y="Rented Bike Count", data=bike, s=10)
```

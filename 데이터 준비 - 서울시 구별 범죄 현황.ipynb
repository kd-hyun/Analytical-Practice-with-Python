## 2000년 ~ 2021년 간, 경찰서별 5대범죄 발생 검거 현황
### 데이터 출처 : https://www.data.go.kr/data/15054738/fileData.do

```python
crime = pd.read_csv("data/경찰서별00-21.csv", encoding = "euc-kr")
```

```python
crime.head()
```

```python
# crime[(crime["죄종"]=="살인")&(crime["발생검거"]=="검거")]
```

```python
for a in range(31):
    crime.loc[a*10 + 0, '죄종결과'] = '살인발생'
    
for a in range(31):
    crime.loc[a*10 + 1, '죄종결과'] = '살인검거'

for a in range(31):
    crime.loc[a*10 + 2, '죄종결과'] = '강도발생'
    
for a in range(31):
    crime.loc[a*10 + 3, '죄종결과'] = '강도검거'
    
for a in range(31):
    crime.loc[a*10 + 4, '죄종결과'] = '강간발생'
    
for a in range(31):
    crime.loc[a*10 + 5, '죄종결과'] = '강간검거'

for a in range(31):
    crime.loc[a*10 + 6, '죄종결과'] = '절도발생'
    
for a in range(31):
    crime.loc[a*10 + 7, '죄종결과'] = '절도검거'
    
for a in range(31):
    crime.loc[a*10 + 8, '죄종결과'] = '폭력발생'
    
for a in range(31):
    crime.loc[a*10 + 9, '죄종결과'] = '폭력검거'    
```
